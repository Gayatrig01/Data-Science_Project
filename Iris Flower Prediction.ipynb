```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
iris = pd.read_csv("C:/Users/HP/Desktop/IRIS.csv")
iris.head()
```

```python
iris.describe()
```

```python
iris.shape
```

```python
iris.info
```

```python
X = iris.drop('species', axis=1)
y = iris['species']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris['species'].unique(), yticklabels=iris['species'].unique())
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
```

```python
plt.subplot(1, 1, 1)
sns.scatterplot(data=X_test, x='sepal_length', y='sepal_width', hue=y_pred, palette='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Flower Classification (Predicted)')

plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 1, 1)
sns.scatterplot(data=X_test, x='petal_length', y='petal_width', hue=y_pred, palette='viridis')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Flower Classification (Predicted)')

plt.tight_layout()
plt.show()
```
